```python
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from collections import OrderedDict
import csv

driver = webdriver.Chrome()
driver.get('https://www.indeed.com/Top-Rated-Workplaces')

csv_file = open('top_workplaces.csv', 'w', encoding='utf-8', newline='')
writer = csv.writer(csv_file)
writer.writerow(['Company','Overall Rating','Rating','Title','Position','Location','Date'])

review = driver.find_elements_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "cmp-rating-description", " " ))]')

links = []
for i in review:
    links.append(i.get_attribute('href'))
links = list(OrderedDict.fromkeys(links))
links.remove(None)




wait = WebDriverWait(driver, .01)
for x in links:
    driver.get(x)
    
    
    company = driver.find_element_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "cmp-company-name", " " ))]').text
    overall_rating = driver.find_element_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "cmp-header-rating-average", " " ))]').text
    index = 1
    while True:
        try:
            print("Scraping Page number " + str(index))
            index = index + 1
            
            
            
            wait_review = WebDriverWait(driver, .01)
            results = wait_review.until(EC.presence_of_all_elements_located((By.XPATH,
                                '//*[contains(concat( " ", @class, " " ), concat( " ", "cmp-review-container", " " ))]')))
    
            for result in results[1:]:
            
                try:
                    rating = result.find_element_by_xpath('.//*[contains(concat( " ", @class, " " ), concat( " ", "cmp-ratingNumber", " " ))]').text
                    title = result.find_element_by_xpath('.//*[contains(concat( " ", @class, " " ), concat( " ", "cmp-review-title", " " ))]//span').text
                    position = result.find_element_by_xpath('.//*[contains(concat( " ", @class, " " ), concat( " ", "cmp-reviewer", " " ))]').text
                    location = result.find_element_by_xpath('.//*[contains(concat( " ", @class, " " ), concat( " ", "cmp-reviewer-job-location", " " ))]').text
                    date = result.find_element_by_xpath('.//*[contains(concat( " ", @class, " " ), concat( " ", "cmp-review-date-created", " " ))]').text
                    
                except Exception as e:
                    print(e)
                    continue
   
                writer.writerow([company,overall_rating,rating,title,position,location,date])
            try:
                next_button = driver.find_element_by_xpath('//a[@data-tn-element="next-page"]')
                next_button.click()
            except Exception as e:
                print(e)
                break
    
        except Exception as e:
            print(e)
            csv_file.close()
            driver.close()
            break
```
